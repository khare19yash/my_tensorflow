```python
import tensorflow as tf
import numpy as np

%matplotlib inline 
import pylab
```

```python
epochs = 4

x_data = np.array([[0,0],[0,1],[1,0],[1,1]])
print(x_data)
```

```python
y_data = np.array([[0],[1],[1],[0]])
y_data.shape
```

```python
pylab.plot(x_data,y_data,'.')
```

```python
x = tf.placeholder(tf.float32, [None,2])
y = tf.placeholder(tf.float32, [None,1])
```

```python
W1 = tf.Variable(tf.random_normal([2,2], stddev = 0.03), name = 'W1')
b1 = tf.Variable(tf.random_normal([2], stddev = 0.03), name = 'b1')


W2 = tf.Variable(tf.random_normal([2,1],stddev = 0.03), name = 'W2')
b2 = tf.Variable(tf.random_normal([1],stddev = 0.03), name = 'b2')
```

```python
hidden_out = tf.add(tf.matmul(x,W1), b1)
hidden_out = tf.nn.relu(hidden_out)
```

```python
y_ = tf.add(tf.matmul(hidden_out,W2),b2)
```

```python
loss = tf.losses.mean_squared_error(y,y_)

optimiser = tf.train.GradientDescentOptimizer(0.1).minimize(loss)
```

```python
init_op = tf.global_variables_initializer()

```

```python
with tf.Session() as sess:
    sess.run(init_op)
    for epoch in range(500):
        avg_cost = 0
        for i in range(4):
            p,c = sess.run([optimiser,loss],
                          feed_dict = {x: x_data, y : y_data})
            avg_cost += c / 4
        print("Epoch:", (epoch+1), "cost =", "{:.3f}".format(avg_cost))
        
```
